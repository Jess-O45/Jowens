Explore variable = Average MRC - Average monthly charges for internet customers in currency - FLOAT Variable

```python
meanMRC = np.mean(df.Average_MRC)
stdMRC = np.std(df['Average_MRC'])
print("The mean value is:", meanMRC)
```

```python
average_mrc_median = df['Average_MRC'].median()
print("Median of Average_MRC:", average_mrc_median)
```

```python
#plot and overlay the standard distribution
plt.figure(figsize=(10, 6))
sns.histplot(df['Average_MRC'], kde=False, bins=10, color='skyblue', edgecolor='black')

xmin, xmax = plt.xlim()  # Get the current limits of the x-axis
x = np.linspace(xmin, xmax, 100)  # Create values from xmin to xmax
p = norm.pdf(x, meanMRC, stdMRC)  # Probability density function for normal distribution

bin_width = (xmax - xmin) / 10  # Bin width (based on number of bins)
scaled_p = p * len(df) * bin_width

# Plot the normal distribution curve
plt.plot(x, scaled_p, 'r-', lw=2, label='Normal Distribution')

plt.title('Standard Distribution of Average_MRC')
plt.xlabel('Average MRC')
plt.ylabel('Frequency')
plt.legend()

# Show the plot
plt.show()
```

The tale is skewed right, meaning the mean is greater than the median. 

```python
# Calculate the PMF (Probability Mass Function)
pmf = df['Average_MRC'].value_counts(normalize=True).sort_index()

# Calculate the CDF (Cumulative Distribution Function)
cdf = pmf.cumsum()

# Display the PMF and CDF
print("PMF (Probability Mass Function):")
print(pmf)

print("\nCDF (Cumulative Distribution Function):")
print(cdf)

# Optional: Plot the PMF and CDF for better visualization
plt.figure(figsize=(12, 6))

# Plot PMF
plt.subplot(1, 2, 1)
plt.bar(pmf.index, pmf.values, color='blue', alpha=0.7)
plt.title('PMF (Probability Mass Function)')
plt.xlabel('Average_MRC')
plt.ylabel('Probability')

# Plot CDF
plt.subplot(1, 2, 2)
plt.step(cdf.index, cdf.values, where='post', color='green', alpha=0.7)
plt.title('CDF (Cumulative Distribution Function)')
plt.xlabel('Average_MRC')
plt.ylabel('Cumulative Probability')

plt.tight_layout()
plt.show()
```

```python
def Cov(xs, ys, meanx=None, meany=None):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    if meanx is None:
        meanx = np.mean(xs)
    if meany is None:
        meany = np.mean(ys)

    cov = np.dot(xs-meanx, ys-meany) / len(xs)
    return cov
```

```python
def Corr(xs, ys):
    xs = np.asarray(xs)
    ys = np.asarray(ys)

    meanx, varx = thinkstats2.MeanVar(xs)
    meany, vary = thinkstats2.MeanVar(ys)

    corr = Cov(xs, ys, meanx, meany) / np.sqrt(varx * vary)
    return corr
```

```python
from thinkstats2 import Mean, MeanVar, Var, Std, Cov

def LeastSquares(xs, ys):
    meanx, varx = MeanVar(xs)
    meany = Mean(ys)

    slope = Cov(xs, ys, meanx, meany) / varx
    inter = meany - slope * meanx

    return inter, slope
```

```python
#sample data 
avg_mrc_array = df['Average_MRC'].to_numpy()

# Step 1: State the hypotheses
# H₀: μ = 125 (mean AVG_MRC is 125)
# H₁: μ ≠ 170 (mean AVG_MRC is not 125)

t_stat, p_value = stats.ttest_1samp(avg_mrc_array, 125)

# Step 3: Interpret the result
alpha = 0.05  # significance level

# Compare p-value with alpha
if p_value < alpha:
    print(f"Reject the null hypothesis (p-value = {p_value:.4f})")
else:
    print(f"Fail to reject the null hypothesis (p-value = {p_value:.4f})")
```
